```python
import numpy as np
import time as time
                        
```

```python
class Node():
    def __init__(self, state, parent):
        self.state = state
        self.parent = parent
        
        self.up = None
        self.left = None
        self.down = None
        self.right = None
    
    

    def cright(self):
        zeroindex = [i[0] for i in np.where(self.state == 0)]
        if zeroindex[1] == 2:
            return False
        else: return True
    def mright(self):
        zeroindex = [i[0] for i in np.where(self.state == 0)]
        right = self.state[zeroindex[0], zeroindex[1]+1]
        tempstate = self.state.copy()
        tempstate[zeroindex[0], zeroindex[1]] = right
        tempstate[zeroindex[0], zeroindex[1] + 1] = 0
        return tempstate
    
    def cup(self):
        zeroindex = [i[0] for i in np.where(self.state == 0)]
        if zeroindex[0] == 0:
            return False
        else: return True
    def mup(self):
        zeroindex = [i[0] for i in np.where(self.state == 0)]
        up = self.state[zeroindex[0]-1, zeroindex[1]]
        finalstate = self.state.copy()
        finalstate[zeroindex[0], zeroindex[1]] = up
        finalstate[zeroindex[0]-1, zeroindex[1]] = 0
        return finalstate
    
    def cleft(self):
        zeroindex = [i[0] for i in np.where(self.state == 0)]
        if zeroindex[1] == 0:
            return False
        else: return True
    
    def mleft(self):
        zeroindex = [i[0] for i in np.where(self.state == 0)]
        left = self.state[zeroindex[0], zeroindex[1]-1]
        finalstate = self.state.copy()
        finalstate[zeroindex[0], zeroindex[1]] = left
        finalstate[zeroindex[0], zeroindex[1]-1] = 0
        return finalstate
    
    def cdown(self):
        zeroindex = [i[0] for i in np.where(self.state == 0)]
        if zeroindex[0] == 2:
            return False
        else: return True
    
    def mdown(self):
        zeroindex = [i[0] for i in np.where(self.state == 0)]
        down = self.state[zeroindex[0]+1, zeroindex[1]]
        finalstate = self.state.copy()
        finalstate[zeroindex[0], zeroindex[1]] = down
        finalstate[zeroindex[0]+1, zeroindex[1]] = 0
        return finalstate
    
    def print_path(self):
        path = [self.state]
        
        while(self.parent):
            self = self.parent      
            path.append(self.state)
            
        step_counter = 0
        while path:
            print('step', step_counter)
            print(path.pop())
            print('\n to \n')
            step_counter += 1
        return step_counter
    
    def BFS(self, goal_state):
        global avgBFS
        global avgstepBFS
        avgBFS = 0
        avgstepBFS = 0
        start = time.time()
        fqueue = [self]
        queue_num_nodes_popped = 0
        queue_max_length = 1
        visited = set([])
        while fqueue:
            if len(fqueue) > queue_max_length:
                queue_max_length = len(fqueue)
                
            curr_node = fqueue.pop(0)
            queue_num_nodes_popped += 1
            visited.add(tuple(curr_node.state.reshape(1,9)[0]))
            
            if np.array_equal(curr_node.state, goal_state):
                avgstepBFS += curr_node.print_path()
                avgBFS += (time.time() - start)
                print('Time spent: %0.2fs' % (time.time() - start))
                return True
            
            else:
                if curr_node.cdown():
                    new_state = curr_node.mdown()
                    
                    if tuple(new_state.reshape(1,9)[0]) not in visited:   
                        curr_node.down = Node(state = new_state, parent= curr_node)
                        fqueue.append(curr_node.down)
                        
                if curr_node.cright():
                    new_state = curr_node.mright()
                    if tuple(new_state.reshape(1,9)[0]) not in visited:
                        curr_node.right = Node(state= new_state, parent= curr_node)
                        fqueue.append(curr_node.right)
                        
                if curr_node.cup():
                    new_state = curr_node.mup()
                    if tuple(new_state.reshape(1,9)[0]) not in visited:
                        curr_node.up = Node(state= new_state, parent=curr_node)          
                        fqueue.append(curr_node.up)
                
                if curr_node.cleft():
                    new_state = curr_node.mleft()
                    if tuple(new_state.reshape(1,9)[0]) not in visited:
                        curr_node.left = Node(state= new_state, parent=curr_node)                  
                        fqueue.append(curr_node.left)
                        
    def IDS(self,goal_state):
        global avgIDS
        global avgstepIDS
        avgIDS = 0
        avgstepIDS = 0
        start = time.time()      
        queue_max_length = 1
        for depth_limit in range(20):
            fqueue = [self]
            depth_queue = [0]
            visited = set([])
            
            while fqueue:
                if len(fqueue) > queue_max_length:
                    queue_max_length = len(fqueue)
                    
                curr_node = fqueue.pop(0)
                current_depth = depth_queue.pop(0)
                
                visited.add(tuple(curr_node.state.reshape(1,9)[0]))
                if np.array_equal(curr_node.state, goal_state):
                    avgstepIDS += curr_node.print_path()
                    print('Time: %0.2fs' % (time.time() - start))
                    avgIDS += (time.time() - start)
                    return True
                else:
                    if current_depth < depth_limit:
                        if curr_node.cdown():
                            new_state = curr_node.mdown()
                            if tuple(new_state.reshape(1,9)[0]) not in visited:   
                                curr_node.down = Node(state = new_state, parent= curr_node)
                                fqueue.append(curr_node.down)
                                depth_queue.insert(0, current_depth+1)
                        if curr_node.cright():
                            new_state = curr_node.mright()
                            if tuple(new_state.reshape(1,9)[0]) not in visited:
                                curr_node.right = Node(state= new_state, parent= curr_node)
                                fqueue.append(curr_node.right)
                                depth_queue.insert(0, current_depth+1)
                        if curr_node.cup():
                            new_state = curr_node.mup()
                            if tuple(new_state.reshape(1,9)[0]) not in visited:
                                curr_node.up = Node(state= new_state, parent=curr_node)          
                                fqueue.append(curr_node.up)
                                depth_queue.insert(0, current_depth+1)
                        if curr_node.cleft():
                            new_state = curr_node.mleft()
                            if tuple(new_state.reshape(1,9)[0]) not in visited:
                                curr_node.left = Node(state= new_state, parent=curr_node)                  
                                fqueue.append(curr_node.left)
                                depth_queue.insert(0, current_depth+1)
                         
```
